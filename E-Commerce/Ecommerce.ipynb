```python
import pandas as pd
data = pd.read_csv('Ecommerce_data.csv')
data['order_date'] = data['order_date'].str.replace('-', '/').apply(lambda x: '-'.join(x.split('/')[::-1]))
data['ship_date'] = data['ship_date'].str.replace('-', '/').apply(lambda x: '-'.join(x.split('/')[::-1]))

data.head()
```

```python
# Convert order_date and ship_date to datetime
data['order_date'] = pd.to_datetime(data['order_date'])
data['ship_date'] = pd.to_datetime(data['ship_date'])
data.head()
```
